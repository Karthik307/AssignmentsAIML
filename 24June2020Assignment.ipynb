```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston
```

```python
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
boston_df
```

```python
# Created a new column with name MEDV and passed the target values from the dictionary
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
# Preparing the Input and Output columns
X = boston_df[['LSTAT']]  # selcting columns a dataframe
y = boston_df['MEDV']
```

```python
boston_df['MEDV']  # selecting column as a series
```

### 1.Consider LSTAT (% lower status of the population) as input and build a SLR Model to predict the MEDV


```python
with plt.style.context("grayscale"):  
    plt.plot(X, y, "bo")  
    plt.xlabel("LSTAT I/P")
    plt.ylabel("MEDV O/P")
    plt.title("Scatter Plot")
```

#### Find the correlation by plotting a scatter plot between those two variables and write down the observations like how the variable are related to each other


```python
with sns.axes_style('dark'):
    sns.regplot(X,y)
```

```python
ax1 = boston_df.plot.scatter(x='LSTAT',y='MEDV',c='r')
```

Looking At the Plot We can observe that as LSTAT(ower status of the population) Increases MEDV(Median value of owner-occupied homes in $1000's) Decreases.And these 2 variables are inversely proptional,and are in moderate negative correlation
