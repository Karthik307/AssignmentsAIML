```python
#1.a.Analyze the  top 5 cities with highest number of female graduates
maxv=df.nlargest(5,'female_graduates')
print(maxv[('name_of_city')])
```

```python
#1.b.Find the 5 cities with lowest number of female graduates and concatenate the two dataframes of top 5 cities with  bottom 5 cities
lowv=df.nsmallest(5,'female_graduates')
print(lowv[('name_of_city')])
```

```python
edu=[maxv,lowv]
r=pd.concat(edu)
r
```
