```python
import pandas as pd
import numpy as n
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo", "bar", "bar", "bar", "bar"],"B": ["one", "one", "one", "two", "two", "one", "one", "two", "two"], "C": ["small", "large", "large", "small","small", "large", "small", "small","large"],"D": [1, 2, 2, 3, 3, 4, 5, 6, 7],"E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                    columns=['C'], aggfunc=n.sum)
```

```python
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                    columns=['C'], aggfunc=n.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                    aggfunc={'D': n.mean,
                             'E': n.mean})
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                    aggfunc={'D': n.mean,
                             'E': [min, max, n.mean]})
table
```
