## Bag of words and ML

```python
#spliting the data --> 80% to training data and 20% testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.CONTENT,data.CLASS,test_size=0.2)b
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Initialising Count Vectorizer
from sklearn.feature_extraction.text import CountVectorizer
#creating object for Count Vectorizer
count_vect=CountVectorizer()
```

```python
#Generate word count for document
word_count_vector=count_vect.fit(X_train)
print(word_count_vector)
```

```python
#to get feature names
word_count_vector.get_feature_names()
```

```python
#Converting to an array
doc_array=word_count_vector.transform(X_train).toarray()
doc_array
```

```python
#Converting to an dataframe
dataf=pd.DataFrame(doc_array,columns=word_count_vector.get_feature_names())
#pd.set_option('display.max_rows',None #for displayig all rows
dataf
```

## Apply Naive Bayes and check the metrics


```python
#TIDF Vectorizer
#import TIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
# Initialize and object for the TFIDF Vectorizer
tfidf = TfidfVectorizer()
```

```python
#apply the tidf vectorizer to the data(X_train)
X_train_transformed = tfidf.fit_transform(X_train)
X_train_transformed
```

```python
X_test_transformed = tfidf.transform(X_test)
X_test_transformed
```

```python
#Feature Names
tfidf.get_feature_names()
```

```python
#Idf of terms
tfidf.idf_
```

```python
#Applying Naive bayes algorithm
from sklearn.naive_bayes import BernoulliNB
#creating object for BerNB
model_BernNB = BernoulliNB()
```

```python
# Applying the Algorithm to the data
# ObjectName.fit(Input, Output)
model_BernNB.fit(X_train_transformed, y_train)
```

```python
#prediction on test data
#Syntax:Objectname.predict(Inputvalues)
y_test_pred = model_BernNB.predict(X_test_transformed)
```

```python
# Compare the actual values(y_test) with predicted values(y_test_pred)
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_test_pred)
```

```python
print(classification_report(y_test, y_test_pred))
```
